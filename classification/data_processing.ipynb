```python
# 3、one-hot处理的feature：性别
df_onehot = pd.get_dummies(df['用户的性别'])
df_onehot = df_onehot.drop(labels=['sex'], axis = 1)
df_onehot
```

```python
# 增加onehot后的列名
s = pd.Series(['保密', '女', '男'])
columns = columns.append(s)
columns
```

```python
dfs_2 = [df, df_onehot]
df = pd.concat(dfs_2, ignore_index = True, axis = 1)
df
```

```python
df.columns = columns
df
```
